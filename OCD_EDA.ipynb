# classes and objects 

```python
class Patient:
    def __init__(self, patient_id, age, gender, ethnicity, marital_status, education_level, ocd_diagnosis_date, duration_of_symptoms, previous_diagnoses, family_history_of_ocd, obsession_type, compulsion_type, ybocs_score_obsessions, ybocs_score_compulsions, depression_diagnosis, anxiety_diagnosis, medications):
        self.patient_id = patient_id
        self.age = age
        self.gender = gender
        self.ethnicity = ethnicity
        self.marital_status = marital_status
        self.education_level = education_level
        self.ocd_diagnosis_date = ocd_diagnosis_date
        self.duration_of_symptoms = duration_of_symptoms
        self.previous_diagnoses = previous_diagnoses
        self.family_history_of_ocd = family_history_of_ocd
        self.obsession_type = obsession_type
        self.compulsion_type = compulsion_type
        self.ybocs_score_obsessions = ybocs_score_obsessions
        self.ybocs_score_compulsions = ybocs_score_compulsions
        self.depression_diagnosis = depression_diagnosis
        self.anxiety_diagnosis = anxiety_diagnosis
        self.medications = medications

    def display_info(self):
        print(f"Patient ID: {self.patient_id}")
        print(f"Age: {self.age}")
        print(f"Gender: {self.gender}")
        print(f"Ethnicity: {self.ethnicity}")
        print(f"Marital Status: {self.marital_status}")
        print(f"Education Level: {self.education_level}")
        print(f"OCD Diagnosis Date: {self.ocd_diagnosis_date}")
        print(f"Duration of Symptoms: {self.duration_of_symptoms} months")
        print(f"Previous Diagnoses: {self.previous_diagnoses}")
        print(f"Family History of OCD: {self.family_history_of_ocd}")
        print(f"Obsession Type: {self.obsession_type}")
        print(f"Compulsion Type: {self.compulsion_type}")
        print(f"Y-BOCS Score (Obsessions): {self.ybocs_score_obsessions}")
        print(f"Y-BOCS Score (Compulsions): {self.ybocs_score_compulsions}")
        print(f"Depression Diagnosis: {self.depression_diagnosis}")
        print(f"Anxiety Diagnosis: {self.anxiety_diagnosis}")
        print(f"Medications: {self.medications}")
        print("")

      # Sample data to create patient object
patients_data = [
    (1018, 32, 'Female', 'African', 'Single', 'Some College', '2016-07-15', 203, 'MDD', 'No', 'Harm-related', 'Checking', 17, 10, 'Yes', 'Yes', 'SNRI'),
    (2406, 69, 'Male', 'African', 'Divorced', 'Some College', '2017-04-28', 180, None, 'Yes', 'Harm-related', 'Washing', 21, 25, 'Yes', 'Yes', 'SSRI'),
    (1188, 57, 'Male', 'Hispanic', 'Divorced', 'College Degree', '2018-02-02', 173, 'MDD', 'No', 'Contamination', 'Checking', 3, 4, 'No', 'No', 'Benzodiazepine'),
     # Add additional patient entries here
]

     # Creating Patient objects
patients = [Patient(*data) for data in patients_data]

     # Display information for all patients
for patient in patients:
    patient.display_info()

     # Calculate average Y-BOCS scores for obsessions and compulsions
average_ybocs_obsessions = sum(patient.ybocs_score_obsessions for patient in patients) / len(patients)
average_ybocs_compulsions = sum(patient.ybocs_score_compulsions for patient in patients) / len(patients)

print(f"Average Y-BOCS Score (Obsessions): {average_ybocs_obsessions}")
print(f"Average Y-BOCS Score (Compulsions): {average_ybocs_compulsions}")

# Filter patients with depression diagnosis
patients_with_depression = [patient for patient in patients if patient.depression_diagnosis == 'Yes']

print("Patients with Depression Diagnosis:")
for patient in patients_with_depression:
    patient.display_info()
```
